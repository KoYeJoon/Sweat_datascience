# 과제 1 lambda 사용

1부터 100까지 3,6,9 게임하기

```
game_369 = ['3','6','9']

clap = lambda x : 'clap ' * (x.count('3') + x.count('6')+x.count('9'))
game_369_check = lambda x : clap(list(str(x))) if any(a for a in list(str(x)) if a in game_369) else str(x)

print(list(map(game_369_check, range(1,101))))
```

## 1) '외국인 비율', '65세 이상 고령자 비율' 계산 후 각각 key(외국인비율, 고령자비율)에 저장

```
import json

with open('/content/drive/MyDrive/Colab Notebooks/SWeat_2020_겨울방학/seoul_people_report','r',encoding='UTF-8-sig') as fr :
  raw_list = [_.strip() for _ in fr.readlines()]
  raw_split = [_.split(',') for _ in raw_list]
```

```
district_dict = {}

for data in raw_split[1:]:
  ds_name = data[0]
  _1,_2,_3,_4,_5 = (int(_) for _ in data[1:])
  district_dict[ds_name] = {'전체' : _1, '남자':_2, '여자' : _3, '외국인' : _4, '고령자' : _5}

for k,v in district_dict.items() :
  district_dict[k]['외국인비율']= (district_dict[k]['외국인'] / district_dict[k]['전체']) * 100
  district_dict[k]['고령자비율']= (district_dict[k]['고령자'] / district_dict[k]['전체']) *100


json_data = json.dumps(district_dict, ensure_ascii=False)

with open('/content/drive/MyDrive/Colab Notebooks/SWeat_2020_겨울방학/1-1_seoul_people_json.json','w',encoding="utf-8") as make_file:
  json.dump(json_data,make_file,ensure_ascii=False)


print(json.dumps(district_dict, ensure_ascii=False,indent=1))
```

## 2) 고령자 비율 15%가 넘고, 외국인 비율이 5% 넘는 구의 인력현황 출력 및 저장

```
F = lambda dataset,k, std : \
      dataset[k] if dataset[k] > std else None

district_list = []
for key,val in district_dict.items() :
  R = F(val,'고령자비율',15)
  if R is not None :
    A = F(val,'외국인비율',5)
    if A is not None :
      district_list.append({key : district_dict[key]})


json_data = json.dumps(district_list, ensure_ascii=False)
with open('/content/drive/MyDrive/Colab Notebooks/SWeat_2020_겨울방학/1-2_seoul_people_output.json','w',encoding="utf-8") as make_file:
  json.dump(json_data,make_file,ensure_ascii=False)

print(json.dumps(district_list, ensure_ascii=False,indent=1))
```

## 3) 서울시 전체 대비 각 구별 인구비율 sorting

```
total = int(raw_split[1][1])
people_percent = {}
rate_for_json = {}
for data in raw_split[2:]:
  ds_name = data[0]
  people_percent[ds_name] = (int(data[1])/total)*100

people_percent = sorted(people_percent.items(),key = lambda x : (x[1]),reverse = True)

for K,V in people_percent :
  rate_for_json[K] = V
  
print(json.dumps(rate_for_json , ensure_ascii=False,indent=1))
```
